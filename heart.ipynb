the following cells represent the comparision of the features with each other graphically

```python
plt.figure(figsize=[15,5])
heart_train.groupby('sex').sex.value_counts()
heart_train[['sex','age']].groupby(['sex'],as_index=False).mean()
sns.barplot(x='sex',y='age',data=heart_train)
plt.show()
```

```python
plt.figure(figsize=[15,5])
heart_train.groupby('sex').sex.value_counts()
heart_train[['sex','chol']].groupby(['sex'],as_index=False).mean()
sns.barplot(x='sex',y='chol',data=heart_train)
plt.show()
```

```python
plt.figure(figsize=[15,5])
heart_train.groupby('target').target.value_counts()
heart_train[['target','sex']].groupby(['target'],as_index=False).mean()
sns.barplot(x='sex',y='target',data=heart_train)
plt.show()
```

```python
fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(131)
ax2=fig.add_subplot(132)
ax3=fig.add_subplot(133)
sns.violinplot(x="sex",y="age",hue="target",data=heart_train,split=True,ax=ax1)
sns.violinplot(x="sex",y="chol",hue="target",data=heart_train,split=True,ax=ax2)
sns.violinplot(x="sex",y="oldpeak",hue="target",data=heart_train,split=True,ax=ax3)
plt.show()
```

```python
total_targeted=heart_train[heart_train['target']==1]
total_not_targeted=heart_train[heart_train['target']==0]
male_targeted=heart_train[heart_train['target']==1&(heart_train['sex']==0)]
male_not_targeted=heart_train[heart_train['target']==0&(heart_train['sex']==0)]
female_targeted=heart_train[heart_train['target']==1&(heart_train['sex']==1)]
female_not_targeted=heart_train[heart_train['target']==0&(heart_train['sex']==1)]
```

```python
plt.figure(figsize=[15,5])
plt.subplot(111)
sns.distplot(total_targeted['age'].dropna().value_counts(),bins=range(0,81,1),kde=False,color='blue')
sns.distplot(total_not_targeted['age'].dropna().value_counts(),bins=range(0,81,1),kde=False,color='red',axlabel='age')
plt.show()
```

```python
plt.figure(figsize=[15,5])
plt.subplot(121)
sns.distplot(female_targeted['age'].dropna().values, bins=range(0, 81, 1), kde=False, color='blue')
sns.distplot(female_not_targeted['age'].dropna().values, bins=range(0, 81, 1), kde=False, color='red', axlabel='Female Age')
plt.show()
```

```python
plt.subplot(122)
sns.distplot(male_targeted['age'].dropna().values, bins=range(0, 81, 1), kde=False, color='blue')
sns.distplot(male_not_targeted['age'].dropna().values, bins=range(0, 81, 1), kde=False, color='red', axlabel='Male Age')
plt.show()
```
